#### Making a basic Bokeh line graph

```python
from bokeh.plotting import figure
from bokeh.io import output_file, show
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]

output_file('Line.html')
f=figure()

f.line(x,y)
show(f)
```

#### Triangle

```python
x=[3,7.5,10]
y=[3,6,9]
 
output_file("Triangle.html")
f=figure()
 
f.triangle(x,y)
show(f)
```

#### Circle

```python
x=[3,7.5,10]
y=[3,6,9]
 
output_file("Circle.html")
f=figure()
 
f.circle(x,y)
show(f)
```

#### Below is a summary of properties which can be added to change the style of the plot:

```python
import pandas
from bokeh.plotting import figure, output_file, show
 
p=figure(plot_width=500,plot_height=400, tools='pan')
 
p.title.text="Cool Data"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"    
 
p.line([1,2,3],[4,5,6])
output_file("Graph.html")
show(p)
```

#### Plotting weather data

```python
import pandas
 
from bokeh.plotting import figure, output_file, show
 
df=pandas.read_excel('files/weather.xlsx')
df["Temperature"]=df["Temperature"]/10
df["Pressure"]=df["Pressure"]/10
 
p=figure(plot_width=500,plot_height=400,tools='pan')
 
p.title.text="Temperature and Air Pressure"
p.title.text_color="Gray"
p.title.text_font="arial"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Temperature (Â°C)"
p.yaxis.axis_label="Pressure (hPa)"    
 
p.circle(df["Temperature"],df["Pressure"],size=0.5)
output_file("Weather.html")
show(p)
```

#### Creating a Time-series plot

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show

df=pd.read_csv('files/adbe.csv', parse_dates=['Date'])

p=figure(width=1200, height=350, x_axis_type='datetime')

p.line(df['Date'], df['Close'], color='orange')

output_file('Time-series.html')
show(p)
```

#### Plotting time intervals from the data generated by the webcam app

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show

df=pd.read_csv('files/time-values.csv', parse_dates=['Start', 'End'])

p=figure(x_axis_type='datetime', height=300, width=900, title='Motion Graph')
p.yaxis.minor_tick_line_color=None
p.ygrid.grid_line_color=None

p.quad(left=df['Start'], right=df['End'], bottom=0, top=1, color='green')

output_file('Webcam.html')
show(p)
```
