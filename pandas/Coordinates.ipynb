## Data Analysis Example: converting addresses to coordinates

O processo de conversão de endereços para as coordenadas é chamado de geocodificação.

```python
import pandas as pd

df1 = pd.read_csv('supermarkets/supermarkets.csv')
df1
```

```python
from geopy.geocoders import ArcGIS
arc = ArcGIS()
```

```python
n=arc.geocode('3995 23rd St, San Francisco, CA 94114')
n
```

```python
df1['Address']=df1['Address'] + ', ' + df1['City'] + ', ' + df1['State'] + ', ' + df1['Country']
df1
```

```python
df1['Coordinates']=df1['Address'].apply(arc.geocode)
df1
```

```python
df1['Latitude']=df1['Coordinates'].apply(lambda x: x.latitude if x != None else None)
df1['Longitude']=df1['Coordinates'].apply(lambda x: x.longitude if x != None else None)
df1
```
